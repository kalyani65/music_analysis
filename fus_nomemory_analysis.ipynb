```python
# Suppose you have a list of genre names
genres = alldatanomemory_df['genre'].unique()

# Suppose you have DataFrame objects for each genre
jazz_datanm_df = alldatanomemory_df[alldatanomemory_df["genre"]=="jazz"]
rap_datanm_df = alldatanomemory_df[alldatanomemory_df["genre"]=="rap"]
rock_datanm_df = alldatanomemory_df[alldatanomemory_df["genre"]=="rock"]
pop_datanm_df = alldatanomemory_df[alldatanomemory_df["genre"]=="pop"]
wc_datanm_df = alldatanomemory_df[alldatanomemory_df["genre"]=="wc"]
ic_datanm_df = alldatanomemory_df[alldatanomemory_df["genre"]=="ic"]

# Create an empty dictionary to store the DataFrame objects
genre_data_dfs = {}

# Assign DataFrame objects to the dictionary with keys corresponding to the genre names
for genre in genres:
    # Dynamically create the variable name
    variable_name = f"{genre}_datanm_df"
    
    # Get the DataFrame object based on the variable name
    datanm_df = locals()[variable_name]
    
    # Assign the DataFrame object to the dictionary with the genre name as the key
    genre_data_dfs[genre] = datanm_df

# Now you can access your DataFrame objects using the genre names as keys in the dictionary
# For example:
display(genre_data_dfs['jazz'])  # Access jazz DataFrame
display(genre_data_dfs['rock'])  # Access rock DataFrame
display(genre_data_dfs['pop'])   # Access pop DataFrame
display(genre_data_dfs['rap'])
display(genre_data_dfs['ic'])
```
